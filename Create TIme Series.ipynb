```python
import pickle
import pandas as pd
import numpy as np
import keras
```

```python
class Hurricane:
    def __init__(self, ocean, id_no, year, name, length, hurr_data ):
        self.ocean = ocean
        self.id_no = id_no
        self.year = year
        self.name = name
        self.length = length
        self.hurr_data = hurr_data
```

```python
def load_hurr_data():
    filename = r'pickledump.txt'
    filehandler = open(filename, 'rb') 
    new_stuff = pickle.load(filehandler)
    return new_stuff
```

```python
hurricanes = load_hurr_data()
```

```python
def create_training_data(hurr_obj, columns, window_length = 6, hurricane_year = 0):
    all_dataframes = []
    for i in range(len(hurricanes)):
        if(int(hurricanes[i].year) > hurricane_year):
            current_df = hurricanes[i].hurr_data
            current_begin = 0
            current_end = current_begin + window_length
            while(current_end <= len(current_df)):
                hold = current_df[current_begin:current_end]
                hold = hold[columns]
                current_begin += 1
                current_end +=1
                all_dataframes.append(hold.reset_index())
    return all_dataframes
```
